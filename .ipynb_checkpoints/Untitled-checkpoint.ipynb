```python
import scipy.io.wavfile as wav
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq
```

```python
# path = 'test.wav'
path = '4_do3.wav'
# path = '1_la3.wav'
```

```python
def format_tableau(sig):
        
        if len(sig.shape) == 2:
            
            signal = (sig[:,0] + sig[:,1])/2
            return np.ravel(signal)

        if len(sig.shape) == 1:
            
            signal = sig.reshape(sig.shape[0], 1)
            return np.ravel(signal)
        
```

```python
rate, signal = wav.read(path)
```

```python
signal.shape
```

```python
sig = format_tableau(signal)
```

```python
sig.shape
```

```python
plt.figure()
# plt.plot(rate[:400, 0])
plt.plot(sig[:200])
plt.show()
```

```python
# sig1 = sig[:,0]
# sig1=sig.copy()
sig.shape[0]
```

```python
X = fft(sig)  # Transformée de fourier
freq = fftfreq(sig.shape[0], d=1/rate)  # Fréquences de la transformée de Fourier
```

```python
freq.shape
```

```python
plt.plot(X[:200])
```

```python
X = np.ravel(X)
X.shape
```

```python
N = sig.shape[0]
# On prend la valeur absolue de l'amplitude uniquement pour les fréquences positives et normalisation
X_abs = np.abs(X[:N//2])*2.0/N
# ne garder que les valeurs positives
freq_pos = freq[:N//2]
```

```python
plt.plot(freq_pos, X_abs, label="Amplitude absolue")
plt.xlim(0, 6000)  # On réduit la plage des fréquences à la zone utile
plt.grid()
plt.xlabel(r"Fréquence (Hz)")
plt.ylabel(r"Amplitude $|X(f)|$")
plt.title("Transformée de Fourier")
plt.show()
```

```python
val = np.max(X_abs)
val
```
